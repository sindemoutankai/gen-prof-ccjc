```python
source_directory = "/Users/yokoyamakai/Desktop/ccjc提出用BU1"
extract_done_dir = "/Users/yokoyamakai/240131_ccjcoutput/extract_done_rating"
good_rating_dir = "/Users/yokoyamakai/240131_ccjcoutput/goodrating"
not_yet_dir = "/Users/yokoyamakai/240131_ccjcoutput/notyet"
```

```python
import os
import shutil
import re

# 各ディレクトリが存在しない場合は作成
for directory in [extract_done_dir, good_rating_dir, not_yet_dir]:
    if not os.path.exists(directory):
        os.makedirs(directory)

# ファイル名の末尾が_番号.xmlのパターン
pattern_rating = re.compile(r'(\d+)_(\d+)\.xml$')

for filename in os.listdir(source_directory):
    source_file = os.path.join(source_directory, filename)
    match = pattern_rating.search(filename)

    if match:
        # ファイル名の末尾に番号がある場合
        rating = int(match.group(2))
        target_file = os.path.join(extract_done_dir, filename)

        # ファイルを移動
        shutil.move(source_file, target_file)
        print(f"Moved: {filename}")

        if rating in [4, 5]:
            # goodrating_1フォルダにコピー
            target_file_good = os.path.join(good_rating_dir, filename)
            shutil.copy(target_file, target_file_good)
            print(f"Copied to goodrating_1: {filename}")

    else:
        # 末尾に番号がない場合
        target_file_notyet = os.path.join(not_yet_dir, filename)
        shutil.move(source_file, target_file_notyet)
        print(f"Moved to notyet_1: {filename}")

print("Completed!")
```

```python
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python
import os
from xml.etree import ElementTree

def extract_occupation(xml_data):
    root = ElementTree.fromstring(xml_data)
    
    # <occupation>タグ以降のみを抽出
    output_elements = []
    occupation_found = False
    for child in root:
        if child.tag == 'personality':
            occupation_found = True
        if occupation_found:
            output_elements.append(ElementTree.tostring(child, encoding='unicode'))
    
    # 抽出した要素を結合して出力
    return "<profile>" + "".join(output_elements) + "</profile>"

def process_xml_file(input_file, output_folder):
    # XMLファイルの読み込み
    with open(input_file, 'r', encoding='utf-8') as file:
        xml_data = file.read()

    # 職業情報の抽出
    extracted_xml = extract_occupation(xml_data)

    # 新しいファイルに書き込み
    output_file = os.path.join(output_folder, os.path.basename(input_file))
    with open(output_file, 'w', encoding='utf-8') as file:
        file.write(extracted_xml)


for filename in os.listdir(input_folder):
    if filename.endswith('.xml'):
        process_xml_file(os.path.join(input_folder, filename), output_folder)
```

```python
from xml.etree import ElementTree

def extract_occupation_only(xml_data):
    root = ElementTree.fromstring(xml_data)
    
    personality_text = ""
    occupation_text = ""

    for child in root:
        if child.tag == 'personality':
            personality_text = ''.join(child.itertext())
        elif child.tag == 'occupation':
            occupation_text = ''.join(child.itertext())

    return personality_text, occupation_text
```

```python
def process_xml_file(input_file):
    # XMLファイルの読み込み
    with open(input_file, 'r', encoding='shift-jis') as file:
        xml_data = file.read()

    # 職業情報の抽出
    personality, occupation = extract_occupation_only(xml_data)
    return personality, occupation, xml_data
    
```
